## Quality-Length Trimming

### Not known what library prep kit the sequencing center used -- don't know adapter sequences 
### For now, I'll filter to reads >75bp and run trimmomatic with standard Nextera PE adapters

#### Length filtering in shell:

````python
for i in *fastq.gz ; do zcat $i | awk 'BEGIN {FS = "\t" ; OFS = "\n"} {header = $0 ; getline seq ; getline qheader ; getline qseq ; if (length(seq) >= 75) {print header, seq, qheader, qseq}}' > "$i".75bp.fastq
````

#### Quality/adapter trim with trimmomatic

````python
while read i ; do java -jar ~/bin/trimmomatic.jar PE -threads 8 "$i".R1.fastq.gz.75bp.fastq "$i".R2.fastq.gz.75bp.fastq "$i".R1.paired.fq.gz "$i".R1.unpaired.fq.gz "$i".R2.paired.fq.gz "$i".R2.unpaired.fq.gz ILLUMINACLIP:/home/ps997/bin/trimmomatic/adapters/NexteraPE-PE.fa:2:30:10 LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 MINLEN:75 ; done < Isoetes_GBS_samples.txt
                                    
````

##### Quality-length trim results
Everything looking good -- eliminating reads < 75 bp drops roughly 5-10% of reads, then ~97% of length-filtered reads pass trimmomatic as paired

##### FASTQC results of quality filtered reads 
Still show nucleotide regularity at first few positions -- R1 files have TGCAG, which was our restriction enzyme cut sequence -- apparently the demultiplexer did not remove this
R2 files tend to end with CGG -- unsure of its origin
